```python
# import library

from Bio.Seq import Seq

# Buat Alignment
from Bio import pairwise2
from Bio.pairwise2 import format_alignment as fa
```

```python
# Declare Seq

seqA = Seq("ATCGAYCGATCGATCAG")
seqB = Seq("AGTCGATCGAT")
```

# Local Alignment

```python
# LOCAL ALIGNMENT
local_align = pairwise2.align.localxx(seqA, seqB)

# * -> unpack operator
# fa -> 5 parameter: seqA, seqB, score, begin, end
for a in local_align:
    print(fa(*a))
```

```python
# global alignment
global_align = pairwise2.align.globalxx(seqA, seqB)

for a in global_align:
    print(fa(*a))
```

```python
# Pake mistach

match = +2
mismatch = -1
gap_open = -0.5
gap_extend = -0.1
global_align_ms = pairwise2.align.globalms(seqA, seqB, match, mismatch, gap_open, gap_extend)

for a in global_align_ms:
    print(fa(*a))
```

# Similarity Analysis

```python
# Hamming Distance

def hamming_distance(seqA, seqB):
    list = []
    for x, y in zip(seqA, seqB):
        if x != y:
            list.append(1)
        else:
            list.append(0)
    return sum(list)

# Panjang harus sama
seq1 = "AGCTAGCGCAGTCATGACTG"
seq2 = "AGCTTGCGCATGCTAGACGA"

hamming_distance(seq1, seq2)
```

# Dotplot

```python
def dotplot(seqA, seqB, nonblank=chr(0x25A0)):
    print(" | " + seqB)
    print("-" * (len(seqB) + 3))
    
    for x in seqA:
        line = ''
        for y in seqB:
            if x == y and x in nonblank:
                line += x
            else:
                line += ' '
        print(x + " | " + line)
dotplot(seq1, seq2)
```
