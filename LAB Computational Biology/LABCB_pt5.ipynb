```python
from Bio import SeqIO

record = SeqIO.read("sequence.gb", "gb")

print(record.id) # Versi unik dari sebuah sequence
print(record.name) # nama pendek dari sebuah seq
print(record.description) # definisi lengkap sekuens (nama bakteri / apapun itu)
print(record.seq) # seq apa aja
print()
for feature in record.features:
    print(feature.type)

print()
print(record.annotations) # informasi tambahan tentang sequence
```
