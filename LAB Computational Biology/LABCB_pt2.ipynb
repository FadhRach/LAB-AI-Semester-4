```python
# GC Content
from Bio.SeqUtils import gc_fraction
from Bio.Seq import Seq
```

```python
seq = Seq("AGCTAGCAGTCAGTCAGTGCAGTCAGTCAGT")
```

# GCAT Composition

```python
# 1. otomatis (panggil function)
gc_composition = gc_fraction(seq)
print("GC Composition:", gc_composition*100, "%")

at_composition = 1 - gc_composition
print("AT Composition:", at_composition*100, "%")
```

```python
# manual gc at
def gc(seq):
    list = []
    for x in seq:
        if x == "G" or x == "C":
            list.append(x)
    return len(list)/len(seq)

print(gc(seq))
print((1-gc(seq)))
```

# Melting Temperature

```python
# melting temperature

from Bio.SeqUtils import MeltingTemp
```

```python
# wallace
mt_wallace = MeltingTemp.Tm_Wallace(seq)
print("Melting Temperature (Wallace):", mt_wallace, "°C")
```

```python
# gc content
mt_gc = MeltingTemp.Tm_GC(seq)
print("Melting Temperature (GC):", mt_gc, "°C")
```

```python
# Nearest Neighbor
mt_at = MeltingTemp.Tm_NN(seq)
print("Melting Temperature (Nearest Neighbor):", mt_at, "°C")
```

```python
# Wallace Manual
def TM_wallace_manual(seq):
    countG = seq.count("G")
    countC = seq.count("C")
    countA = seq.count("A")
    countT = seq.count("T")
    return 2 * (countA + countT) + 4 * (countG + countC)

print("Melting Temperature (Wallace Manual):", TM_wallace_manual(seq), "°C")
```

# LATSOL DNA COMPOSITION ANALYSIS (word)

```python
sequenceA = Seq("AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC")
sequenceB = Seq("CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG")
```

```python
# Determine the molecular weight of both sequnece.
print("Molecular Weight of Sequence A:", molecular_weight(sequenceA), "g/mol")
print("Molecular Weight of Sequence B:", molecular_weight(sequenceB), "g/mol")
```

```python
# Refer to the Wallace rule to calculate the melting point of sequene A and sequence B and plot them using Matplotlib library.
import matplotlib.pyplot as plt

# wallace
mt_wallace = MeltingTemp.Tm_Wallace(sequenceA)
print("Melting Temperature (Wallace) for Sequence A:", mt_wallace, "°C")

mt_wallace_B = MeltingTemp.Tm_Wallace(sequenceB)
print("Melting Temperature (Wallace) for Sequence B:", mt_wallace_B, "°C")

# Plotting
labels = ['Sequence A', 'Sequence B']
melting_temps = [mt_wallace, mt_wallace_B]
plt.bar(labels, melting_temps, color=['blue', 'red'])
plt.ylabel('Melting Temperature (°C)')
plt.title('Melting Temperature of Sequences A and B (Wallace Rule)')
```

```python
# Calculate the GC content of sequence A and sequence B

gc_content_A = gc_fraction(sequenceA)
gc_content_B = gc_fraction(sequenceB)

print("GC Content of Sequence A:", gc_content_A*100, "%")
print("GC Content of Sequence B:", gc_content_B*100, "%")
```

```python
# Calculate the AT content of sequence A and sequence B

at_content_A = 1 - gc_content_A
at_content_B = 1 - gc_content_B

print("AT Content of Sequence A:", at_content_A*100, "%")
print("AT Content of Sequence B:", at_content_B*100, "%")
```

```python
# Plot the GC and AT content of sequence A and sequence B using Matplotlib library.
labels = ['Sequence A', 'Sequence B']
gc_contents = [gc_content_A*100, gc_content_B*100]
at_contents = [at_content_A*100, at_content_B*100]
x = range(len(labels))
plt.bar(x, gc_contents, width=0.4, label='GC Content', color='blue', align='center')
plt.bar(x, at_contents, width=0.4, label='AT Content', color='red', align='edge')
plt.xticks(x, labels)
plt.ylabel('Content (%)')
plt.title('GC and AT Content of Sequences A and B')
plt.legend()
plt.show()
```
