```python
from Bio import SeqIO

seqA = SeqIO.read("a.fasta", "fasta")
seqB = SeqIO.read("b.fasta", "fasta")
```

```python
# 1. Sequence manipulation
# a. Cari dan display total panjang dari kedua sequence
len_a = len(seqA)
len_b = len(seqB)
total = len_a + len_b
print("total sequence a dan b adalah", total)

# b. Find and display which index the first time codon (TAC) appears in both Sequence A and Sequence B
print(seqA.seq.index('TAC'))
print(seqB.seq.index('TAC'))

# c. Take the first 10 nucleotides of Sequence A and the last 15 nucleotides of Sequence B, then combine the sequences and reverse it to a new sequence C
ten_first = seqA.seq[:10]
fifteen_last = seqB.seq[-15:]
combine = ten_first + fifteen_last
seqC = combine[::-1]
seqC
```

```python
# 2. Sequence analysis & Plotting

# a. Determine and display the nucleotides base (A,C,G,T) frequency for Sequence A, B, and C with a chart using matplotlib library
import matplotlib.pyplot as plt
from collections import Counter
counter_a = Counter(seqA.seq)


# b. Determine and display the GC&AT content percentage for Sequence A, B, and C
from Bio.SeqUtils import gc_fraction
gc_a = gc_fraction(seqA) * 100
at_a = 100 - gc_a
print(f"Percentage of GC content in Sequence A: {gc_a}%")
print(f"Percentage of AT content in Sequence A: {at_a}%")

# c. Determine and display the melting temperature for Sequence A,B and C (Using built in Wallace Method)
from Bio.SeqUtils import MeltingTemp
print(MeltingTemp.Tm_Wallace(seqA.seq, strict=False))
print(MeltingTemp.Tm_Wallace(seqB.seq, strict=False))
print(MeltingTemp.Tm_Wallace(seqC, strict=False))

# d. Determine and display the molecular weight for Sequence A, B, and C
from Bio.SeqUtils import molecular_weight
print(molecular_weight(seqA))
print(molecular_weight(seqB))
print(molecular_weight(seqC))
```
