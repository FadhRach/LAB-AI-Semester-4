```python
from Bio.Seq import Seq
```

```python
seqA = Seq("ATGGCAATCTAA")
seqA
```

```python
# Complement -> kebalikan pasangan dari dna nya (G-C, A-T)
seqA_Comp = seqA.complement()
print("Seq Normal   :", seqA)
print("Complemeent  :", seqA_Comp)


# Antisense strand -> mengamnalisis dan moenconvert menjadi sebuah RNA
seqA_antisense = seqA.reverse_complement()
print("Anti Sense   :", seqA_antisense)
```

```python
RNA = seqA.transcribe()
print("DNA  :", seqA)
print("RNA  :", RNA)
```

```python
aminoAcid = RNA.translate()
aminoAcid
# translate -> mengambil hurf depan dari amino acid saja
# * -> stop kodon
```

```python
from Bio.SeqUtils import seq3, seq1

polypeptide = seq3(aminoAcid)
print(polypeptide)
polypeptide_1 = seq1(polypeptide)
polypeptide_1
```
