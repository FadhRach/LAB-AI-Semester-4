```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split

# Membuat dataset dengan 10 fitur, tapi hanya 2 fitur pertama yang penting
# 8 fitur sisanya adalah noise (sampah) yang memicu overfitting
np.random.seed(42)
X = np.random.rand(100, 10) 
y = 3 * X[:, 0] + 1.5 * X[:, 1] + np.random.randn(100) * 0.5 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 1. Regresi Linear Biasa (Tanpa Regularization / Overfit)
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)

# 2. Ridge Regression (L2 Regularization) -> Alpha = Lambda
model_ridge = Ridge(alpha=1.0)
model_ridge.fit(X_train, y_train)

# 3. Lasso Regression (L1 Regularization) -> Alpha = Lambda
model_lasso = Lasso(alpha=0.1) # Alpha disesuaikan agar terlihat efeknya
model_lasso.fit(X_train, y_train)

# Membuat DataFrame untuk membandingkan Bobot/Koefisien tiap fitur
df_coef = pd.DataFrame({
    'Fitur': [f'Fitur_{i+1}' for i in range(10)],
    'Koefisien Linear (Asli)': model_lr.coef_,
    'Koefisien Ridge (L2)': model_ridge.coef_,
    'Koefisien Lasso (L1)': model_lasso.coef_
})

print("=== PERBANDINGAN KOEFISIEN (DAMPAK REGULARIZATION) ===")
display(df_coef.round(3))

# ANALISIS UNTUK DOSEN:
# 1. Pada Linear Regression, fitur 3-10 (noise) memiliki nilai koefisien yang cukup besar.
# 2. Pada Ridge (L2), nilai koefisien ditekan menjadi lebih kecil/mendekati nol.
# 3. Pada Lasso (L1), fitur noise BANYAK YANG DITEKAN TEPAT MENJADI 0.000! 
#    Ini membuktikan teori bahwa L1 mengurangi kompleksitas model dengan mengeleminasi fitur secara otomatis.
```
