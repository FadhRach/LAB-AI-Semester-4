```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# 1. Membuat Item Profile (Karakteristik dari masing-masing lagu)
# Angka 1 berarti lagu tersebut memiliki fitur itu, angka 0 berarti tidak.
# Bisa juga menggunakan angka kontinyu (0.0 sampai 1.0)
lagu_features = {
    'Acoustic': [1, 1, 0, 1, 0],
    'Piano_Driven': [1, 0, 1, 1, 0],
    'Upbeat_Tempo': [0, 0, 0, 0, 1],
    'Vocal_Focus': [1, 1, 1, 1, 0]
}

lagu_titles = [
    'Supermarket Flowers', 
    'Photograph', 
    'All of Me', 
    'Someone Like You', 
    'Heavy Metal Rock' # Lagu jebakan
]

df_items = pd.DataFrame(lagu_features, index=lagu_titles)

print("=== ITEM PROFILE (MATRIKS FITUR LAGU) ===")
display(df_items)
```

```python
# =======================================================
# 2. MEMBANGUN USER PROFILE & PENALARAN REKOMENDASI
# =======================================================

# Skenario: "User Target" sangat menyukai lagu 'Supermarket Flowers' dan 'Photograph'.
# Kita asumsikan dia memberi rating positif, sehingga profilnya adalah rata-rata fitur kedua lagu itu.
lagu_disukai = ['Supermarket Flowers', 'Photograph']

# Mengambil fitur dari lagu yang disukai lalu mencari rata-ratanya
user_profile = df_items.loc[lagu_disukai].mean()

# Ubah ke dalam bentuk DataFrame / Vektor 2D agar bisa dihitung oleh sklearn
df_user_profile = pd.DataFrame([user_profile], index=['Profil_Selera_User'])

print("\n=== USER PROFILE (Berdasarkan lagu yang disukai) ===")
display(df_user_profile)
```

```python
# =======================================================
# 3. MENGHITUNG COSINE SIMILARITY ANTARA USER DAN LAGU LAIN
# =======================================================

# Lagu yang BELUM didengar oleh User Target
lagu_belum_didengar = ['All of Me', 'Someone Like You', 'Heavy Metal Rock']
df_unseen_items = df_items.loc[lagu_belum_didengar]

# Menghitung kemiripan (Cosine Similarity)
similarity_scores = cosine_similarity(df_user_profile, df_unseen_items)

# Tampilkan hasil
df_recommendation = pd.DataFrame(
    similarity_scores.T, 
    index=lagu_belum_didengar, 
    columns=['Skor_Kecocokan']
).sort_values(by='Skor_Kecocokan', ascending=False)

print("\n=== HASIL REKOMENDASI (CONTENT-BASED) ===")
display(df_recommendation.round(3))

# ANALISIS UNTUK DOSEN:
# Sistem merekomendasikan 'Someone Like You' di urutan pertama (Skor 0.866) 
# karena fiturnya sangat mirip dengan 'Supermarket Flowers' dan 'Photograph' (Akustik & Vokal kuat).
# Sistem membuang 'Heavy Metal Rock' (Skor 0.0) TANPA perlu melihat apakah user lain menyukainya.
```
