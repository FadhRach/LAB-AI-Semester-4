```python
# Liked distribution

import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
mode_dist = df['liked'].value_counts(normalize=True)
plt.pie(mode_dist, labels=['Liked (1)', "Dislike (0)"], autopct='%1.1f%%')
plt.title("Liked and Disliked Distribution")
plt.show()
```

```python
columns = df.drop(columns=['liked']).columns

for column in columns:
    plt.figure(figsize=(8, 6))

    # Disliked
    plt.hist(df[df['liked'] == 0][column],
            bins=30, alpha=0.7, label='Disliked')

    # Liked
    plt.hist(df[df['liked'] == 1][column],
            bins=30, alpha=0.7, label='Liked')

    plt.title(f"Histogram of {column} by liked status")
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True)
    plt.show()
```
