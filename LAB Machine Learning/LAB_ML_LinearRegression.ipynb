# Data Preprocessing

```python
# Drop Missing Values > 50%

thresshold = 0.5
missing_ratio = train_data.isnull().mean()
cols_to_drop = missing_ratio[missing_ratio > thresshold].index
train_data = train_data.drop(cols_to_drop, axis=1)
```

```python
train_data.shape
```

```python
# Imputation for Missing Values
for col in train_data.columns:
    if train_data[col].dtype == 'object':
      mode_val = train_data[col].mode()[0]
      train_data[col] = train_data[col].fillna
    else:
      median_val = train_data[col].median()
      train_data[col] = train_data[col].fillna
```

```python
# Re-Check Values in Training Data
train_data.isnull().sum()
```

```python
train_data.head()
```

```python
# Drop column ID
train_data = train_data.drop(columns='Id')
```

```python
# Identify Categorical Columns

cat_cols = train_data.select_dtypes(include=['object']).columns
cat_cols
```

```python
# Label Encoding

from sklearn.preprocessing import LabelEncoder

label_encoders = {}

for col in cat_cols:
  le = LabelEncoder()
  train_data[col] = le.fit_transform(train_data[col])
```

```python
# Seperate Target Class

y = train_data['SalePrice']
x = train_data.drop('SalePrice', axis=1)
```

```python
# Data Transformation

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
# Split Data into Training and Validation with Ratio 70:30

from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(x_scaled, y, test_size=0.3, random_state=42)
```

```python
train_data.head()
```

# Training Linear Regression with OLS

```python
from sklearn.linear_model import LinearRegression

LR_OLS = LinearRegression()
LR_OLS.fit(x_train, y_train)
```

```python
# Get the model Parameters
print('Intercept(Bias): ', LR_OLS.intercept_)
print('Weight: ', LR_OLS.coef_)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np

pred = LR_OLS.predict(x_val)

rmse = np.sqrt(mean_squared_error(y_val, pred))
mae = mean_absolute_error(y_val, pred)
r2 = r2_score(y_val, pred)

print('RMSE: ', rmse)
print('MAE: ', mae)
print('R2: ', r2)
```
