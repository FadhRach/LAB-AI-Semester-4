```python
import cv2
import os
# euclidean -> untuk cari nilai similarity index
from scipy.spatial.distance import euclidean
```

```python
# bikin flat histogram dari image library
IMAGE_DIR = 'dataset_PT9/image_library'
TEST_DIR = 'dataset_PT9/image_test'
features = []
# [[images_name_a, flat_hist_a],
#  [images_name_b, flat_hist_b],...
# ]
for filename in os.listdir(IMAGE_DIR):
    # image.jpg -> ['image', 'jpg']
    image_name = filename.split('.')[0]
    image_bgr = cv2.imread(os.path.join(IMAGE_DIR, filename))
    image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

    # calcHist([image],[jum_channel],[bin],[ranges])
    # channel -> rgb -> 3 channel -> 0,1,2
    # 1 channel -> 8 bin = kombinasi tiap channel warna
    # range -> range di masing2 bin
    hist = cv2.calcHist([image_bgr], [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256])
    
    # normalized
    # image high quality > image low quality
    normalized_hist = cv2.normalize(hist, None)

    # flat histogram -> 3d -> 1d array
    flat_hist = normalized_hist.flatten()

    features.append((image_name, flat_hist))
```

```python
# buat histogram utk query image
query_img_bgr = cv2.imread(TEST_DIR + '/miaw.jpeg')
query_img_rgb = cv2.cvtColor(query_img_bgr, cv2.COLOR_BGR2RGB)
query_hist = cv2.calcHist([query_img_bgr], [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256])

normalized_query_hist = cv2.normalize(query_hist, None)

flat_query_hist = normalized_query_hist.flatten()
```

```python
# Cari similarity index
results = []
for name, hist in features:
    distance = euclidean(flat_query_hist, hist)
    results.append((distance, name))
```

```python
results = sorted(results)
for distance, name in results:
    print(f'{name} : {distance:.2f}')
```
