## PLOTTING DAN HISTOGRAM EQUALIZATION

# PLOT

```python
import matplotlib.pyplot as plt

x = [1,2,3]
y = [4,5,6]

plt.figure(figsize=(10,4))

# subplot(row, col, idx)
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.title("Sample Plot x, y")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.subplot(1, 2, 2)
plt.plot(y, x, 'g.-')
plt.title("Sample Plot y, x")
plt.xlabel("Y-axis")
plt.ylabel("X-axis")
plt.show()
```

# HISTOGRAM EQUALIZATION

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('image.jpg')
# print(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# cv2.imshow('test', gray)
# cv2.waitKey(0)

height, width, _ = img.shape

intensity_counter = np.zeros(256, dtype=int)
for i in range(height):
    for j in range(width):
        intensity_counter[gray[i][j]] += 1

equalized_img = cv2.equalizeHist(gray)
equalized_counter = np.zeros(256, dtype=int)
for i in range(height):
    for j in range(width):
        equalized_counter[equalized_img[i][j]] += 1

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(intensity_counter)
plt.xlabel('quantity')
plt.ylabel('intensity')
plt.title('Before Equalization')

plt.subplot(1,2,2)
plt.plot(equalized_counter)
plt.xlabel('quantity')
plt.ylabel('intensity')
plt.title('After Equalization')
plt.show()
```

```python
results = np.hstack((gray, equalized_img))
cv2.imshow('Gray vs Equalized', results)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
