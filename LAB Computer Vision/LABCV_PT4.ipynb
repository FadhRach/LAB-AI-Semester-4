FEATURE EXTRACTION ~ EDGE
- [] Sobel: Turunan ke-1
- [] Laplacian
- [] Canny: Paling bagus, cuma lambat dan kompleks

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('train_369.jpg', cv2.IMREAD_GRAYSCALE)
print(img)
```

```python
# Sobel Operator
# Sobel-X -> Sobel(image, depth, dx, dy, ksize) 
# depth untuk formatting hasil angka setelah di turunin
# dx untuk arah x, dy untuk arah y, ksize untuk ukuran kernel
sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, 3)

# Sobel-Y -> Sobel(image, depth, dx, dy, ksize)
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, 3)

# Result
soble_res = np.sqrt(sobel_x**2 + sobel_y**2)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img, 'gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(soble_res, 'gray')
plt.title('Sobel Result')
plt.axis('off')

plt.show()
```

```python
# Laplacian -> bisa detect semua arah
# Laplacian(image, depth, ksize)
laplacian_res = cv2.Laplacian(img, cv2.CV_64F, ksize=3)
laplacian_res = cv2.convertScaleAbs(laplacian_res) # untuk mengubah hasil ke dalam bentuk uint8

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img, 'gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(laplacian_res, 'gray')
plt.title('Laplacian Result')
plt.axis('off')

plt.show()
```

```python
# Canny
# 1. Noise Reduction
# 2. Calculate intensity gradient
# 3. Supress False Edges
# 4. Thresholding

# Canny(image, lower_threshold, upper_threshold)
canny_res = cv2.Canny(img, 100, 200)
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img, 'gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(canny_res, 'gray')
plt.title('Canny Result')
plt.axis('off')
plt.show()
```

```python
# All Results
plt.figure(figsize=(12,20))
plt.subplot(1,4,1)
plt.imshow(img, 'gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1,4,2)
plt.imshow(soble_res, 'gray')
plt.title('Soble Result')
plt.axis('off')
plt.subplot(1,4,3)
plt.imshow(laplacian_res, 'gray')
plt.title('Laplacian Result')
plt.axis('off')
plt.subplot(1,4,4)
plt.imshow(canny_res, 'gray')
plt.title('Canny Result')
plt.axis('off')
plt.show()
```
