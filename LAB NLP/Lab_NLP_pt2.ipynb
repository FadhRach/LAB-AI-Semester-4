#Stemming

```python
#Stemming

#1. PorterStemmer -> algoritma lambat, tapi secara hasil cukup akurat (only english)
#2. SnowballStemmer -> sama seperti porter tapi versi lebih bagus dan support multi languange
#3. LancasterStemmer -> algoritmanya cepat, tapi hasilnya buruk (akurasinya buruk)

import nltk
from nltk.stem import PorterStemmer
from nltk.stem import SnowballStemmer
from nltk.stem import LancasterStemmer

porter = PorterStemmer()
snowball = SnowballStemmer('english')
lancaster = LancasterStemmer()
```

```python
word_list = ['dancing','changing','drawn','moved']

for word in word_list:
  print(f'Porter= {word} -> {porter.stem(word)}')
  print(f'Snowball= {word} -> {snowball.stem(word)}')
  print(f'Lancaster= {word} -> {lancaster.stem(word)}\n')
```

```python
# Lemmatization (Lmematiziing)

from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemma = WordNetLemmatizer()
```

```python
word = 'better'

# Kegunaan lemmatizing lainnya -> bisa mengubah sebuah bentuk kata menjadi bentuk kata lainnya,
# better -> good, verb -> adberb, adjective -> verb, dsb
# a -> adjective
adjective = lemma.lemmatize(word, 'a')
# r -> adverb
adverb = lemma.lemmatize(word, 'r')
# v -> verb
verb = lemma.lemmatize(word,'v')
# n -> noun
noun = lemma.lemmatize(word, 'n') # tidak pakai n juga takpaa

print(word)
print("Adjective: " + adjective)
print("Adverb: " + adverb)
print("Verb: " + verb)
print("Noun: " + noun)
```

```python
# Mencari sinonim dan antonim dari sebuah kata

from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
sentence = 'Quickly kangaroos caused chaos'

word_list = word_tokenize(sentence)

# print(word_list)

for word in word_list:
  synonyms = []
  antonyms = []

  synsets = wordnet.synsets(word)

  for syn in synsets:
    for lemma in syn.lemmas():
      synonyms.append(lemma.name())
      for ant in lemma.antonyms():
        antonyms.append(ant.name())

  print(word)

  if synonyms:
    print('Synonyms: ' + str(set(synonyms[:5])))
  else:
    print('No synonyms')

  if antonyms:
    print('Antonyms: ' + str(set(antonyms[:5])))
  else:
    print('No antonyms')

  print()
```
