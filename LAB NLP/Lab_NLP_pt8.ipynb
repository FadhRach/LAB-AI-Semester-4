# N-Gram (Berdampingan dalam sebuah text)

contoh kalimat: "The quick brown fox"

1. Unigram : "the", "quick", "brown", "fox"
2. Bigram : "the quick", "quick brown", "brown fox"

```python
# Import

import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# buat class model N-Gram

class NGramModel:
    # constructor
    def __init__(self, n):
        self.n = n
        self.vectorizer = CountVectorizer(analyzer='word',ngram_range=(n, n))

    def fit_transform(self, corpus):
        return self.vectorizer.fit_transform(corpus)
    
    def transform(self, corpus):
        return self.vectorizer.transform(corpus)
```

```python
# Document
document = [
    'She sells sea shells by the sea shore',
    'The shells she sells are surely sea shells',
    'So if she sells sea shells on the sea shore',
]

query = 'surely she sells sea shells'
```

```python
ngram_names = {1: 'unigram', 2: 'bigram', 3: 'trigram'}

# looping untuk n = 1, 2, 3
for n in [1, 2, 3]:
    # buat model n-gram
    ngram_model = NGramModel(n)
    
    # fit dan transform data dokumen
    target_vectors = ngram_model.fit_transform(document)
    
    # transform query
    query_vector = ngram_model.transform([query])
    
    # hitung cosine similarity antara query dan dokumen
    similarities = cosine_similarity(query_vector, target_vectors)[0]
    
    # tambahkan similarity ke dataframe result
    column_name = 'similarity_' + ngram_names[n]

    result_df[column_name] = similarities
```
