```python
from nltk.tokenize import word_tokenize
from nltk.parse import ChartParser
from nltk import CFG
```

# Grammar Parsing With NLTK

```python
def extract_information(parse_tree):
    for subtree in parse_tree.subtrees():
        if subtree.label() == 'NP':
            print("Noun Phrase:", ' '.join(subtree.leaves()))
```

```python
def demonstrate_parsing(sentences, grammar):
    words = word_tokenize(sentences)
    parser = ChartParser(grammar)
    try:
        parses = list(parser.parse(words))
        if parses:
            for tree in parser.parse(words):
                print(tree, '\n')

                tree.pretty_print()

                print("Extracted Information:")
                extract_information(parses[0])
    except:
        print("Error during parsing")
```

```python
grammar = CFG.fromstring('''
                         S -> NP VP
                         NP -> Det N | 'i'
                         VP -> V NP | V NP PP
                         PP -> P NP
                         Det -> 'the' | 'a'
                         N -> 'snail' | 'snake' | 'stick' | 'sleigh'
                         V -> 'slide' | 'saw' | 'slid' | 'spoil'
                         P -> 'in' | 'with'
                         ''')

sentences = [
    'the snail slide the snake',
    'i saw a stick',
    'the snail slid the stick in the sleigh',
    'i spoil the snail with the stick',
]

for i, sentences in enumerate(sentences):
    print(f"Parsing sentence {i+1}: '{sentences}'")
    demonstrate_parsing(sentences, grammar)
    print("\n")
```

```python
# Dependency Parsing with NLTK
import spacy
```

```python
def parse_dependency_tree(sentence):
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(sentence)
    
    print('Formatted Dependency Parse Tree:')
    for token in doc:
        print(f"{token.text} --> {token.dep_} --> {token.head.text} {token.pos_}")
        # token.text: kata yang sedang diproses
        # token.dep_: jenis dependency antara token dan head-nya
        # token.head.text: kata rootnya
        # token.pos_: part of speech dari token tersebut
```

```python
sentence = "Elon musk founded SpaceX, and the headquarters are in Palo Alto"

parse_dependency_tree(sentence)
```

# Named Entity Recognition (NER) with SpaCy

```python
import spacy
```

```python
def extract_ner(sentence):
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(sentence)

    # doc.ents
    named_entities = {
        'person' : [ent.text for ent in doc.ents if ent.label_ == 'PERSON'],
        'organization' : [ent.text for ent in doc.ents if ent.label_ == 'ORG'],
        'location' : [ent.text for ent in doc.ents if ent.label_ == 'GPE']
    }
    return named_entities
```

```python
sentence = "Elon Musk founded Tesla, and the headquarters are in Texas"

named_entities = extract_ner(sentence)

print("Extracted Named Entities:")
print("Persons:", named_entities['person'])
print("Organizations:", named_entities['organization'])
print("Locations:", named_entities['location'])
```
